# 1. Import libraries

```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('seaborn-darkgrid')
```

# 2. Put payoff

```python
def put_payoff(St, strike_price, premium):
    return np.where(St < strike_price, strike_price - St, 0) - premium 
```

# 3. Define parameters

```python
# Bitcoin price 
spot_price = 17100 

# Long put
strike_price_long_put = 16700 
premium_long_put = 285

# Short put
strike_price_short_put = 16300 
premium_short_put = 190

# Bitcoin price range at expiration of the put
St = np.arange(0.9*spot_price,1.05*spot_price,1) 
```

# 4. Long 16700 strike put payoff

```python
payoff_long_put = put_payoff(St, strike_price_long_put, premium_long_put)
# Plot
fig, ax = plt.subplots(figsize=(8,5))
ax.spines['bottom'].set_position('zero')
ax.plot(St,payoff_long_put,label='Long 16700 Strike Put',color='g')
plt.xlabel('Bitcoin Stock Price')
plt.ylabel('Profit and loss')
plt.legend()
plt.show()
```

# 5. Short 16300 strike put payoff

```python
payoff_short_put = put_payoff(St, strike_price_short_put, premium_short_put) * -1.0
# Plot
fig, ax = plt.subplots(figsize=(8,5))
ax.spines['bottom'].set_position('zero')
ax.plot(St,payoff_short_put,label='Short 16300 Strike Put',color='r')
plt.xlabel('Bitcoin Price')
plt.ylabel('Profit and loss')
plt.legend()
plt.show()
```

# 6. Bear put spread payoff

```python
payoff_bear_put_spread = payoff_long_put + payoff_short_put

print ("Max Profit:", max(payoff_bear_put_spread))
print ("Max Loss:", min(payoff_bear_put_spread))

# Plot
fig, ax = plt.subplots(figsize=(8,5))
ax.spines['bottom'].set_position('zero')
ax.plot(St,payoff_long_put,'--',label='Long 16700 Strike Put',color='g')
ax.plot(St,payoff_short_put,'--',label='Short 16300 Strike Put',color='r')
ax.plot(St,payoff_bear_put_spread,label='Bear Put Spread')
plt.xlabel('Bitcoin Price')
plt.ylabel('Profit and loss')
plt.legend()
plt.show()
```
